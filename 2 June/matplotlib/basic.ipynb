```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# plot

```python
X_point = [1,2,3,4,5]
Y_point = [1,4,9,16,25]
plt.plot(X_point,Y_point)
plt.show
```

```python
X_point = [0,6]
Y_point = [0,50]
plt.plot(X_point,Y_point,'o')  ## o is used to create scatter plot
plt.show()
```

```python
X_point = [1,2,3,4,5]
plt.plot(X_point,marker = 'o') ## marker is used to create line brtween two points 
plt.show
```

```python
X_point = np.array([1,2,3,4,5])
plt.plot(X_point,marker = '*') ## marker is used to create line brtween two points 
plt.show
```

```python
X_point = np.array([1,2,3,4,5])
plt.plot(X_point,'o:b')         ## o is used to create scatter plot and :b is used to create blank line
```

```python
X_point = np.array([1,2,3,4,5])
plt.plot(X_point,marker = 'o',ms = 30)
```

```python
X_point = np.array([1,2,3,4,5])
plt.plot(X_point,marker = '*',ms = 30,mec='r')
```

```python
X_point = np.array([1,2,3,4,5])
plt.plot(X_point,marker = '*',ms = 30,mec='r',mfc='g')
```

```python
X_point = np.array([1,2,3,4,5])
plt.plot(X_point,ls='dotted')
plt.show  
```

```python
X_point = np.array([1,2,3,4,5])
plt.plot(X_point,ls='dashed')
plt.show  
```

```python
X_point = np.array([1,2,3,4,5])
plt.plot(X_point,linewidth= 5)
plt.show 
```

```python
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,3,2,1])
plt.plot(X_point)
plt.plot(y_point)
plt.show
```

```python
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,3,2,1])
plt.xlabel("x data points")
plt.ylabel("y data points")
plt.plot(X_point)
plt.plot(y_point)
plt.show
```

```python
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,3,2,1])
plt.title("x and y data points")
plt.xlabel("x data points")
plt.ylabel("y data points")
plt.plot(X_point)
plt.plot(y_point)
plt.show
```

```python
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,3,2,1])
plt.title("x and y data points")
plt.xlabel("x data points")
plt.ylabel("y data points")
plt.grid()
plt.plot(X_point)
plt.plot(y_point)
plt.show
```

```python
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,3,2,1])
plt.subplot(1,2,1)
plt.plot(X_point,y_point)
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,23,23,11])
plt.subplot(1,2,2)
plt.plot(X_point,y_point)
plt.show
```

## Scatter plot

```python
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,3,2,1])
plt.scatter(X_point,y_point)
plt.show
```

```python
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,3,2,1])
plt.scatter(X_point,y_point)

X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,23,23,11])
plt.scatter(X_point,y_point)
plt.show
```

```python
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,3,2,1])
colour = np.array(['red','green','blue','yellow','black'])
plt.scatter(X_point,y_point,c = colour,alpha=0.5)
```

```python
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,3,2,1])
colour = np.array(['red','green','blue','yellow','black'])
size = np.array([100,200,300,400,500])
plt.scatter(X_point,y_point,c = colour,alpha=0.5,s=size)
```

```python
X_point = np.array([1,2,3,4,5])
y_point = np.array([5,4,3,2,1])
colour = np.array(['red','green','blue','yellow','black'])
size = np.array([100,200,300,400,500])
plt.scatter(X_point,y_point,c = colour,alpha=0.5,s=size,cmap='coolwarm')
plt.colorbar()
plt.show
```
